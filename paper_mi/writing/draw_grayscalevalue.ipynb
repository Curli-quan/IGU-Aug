```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import cv2
from skimage.exposure import histogram
from sklearn.metrics import mutual_info_score
from einops import rearrange
import seaborn as sns
import pandas as pd

img = cv2.imread("./384_3167.jpg")
img.shape

patch = img[100:164, 100:164, 0]
patch_seq = rearrange(patch, "h w -> (h w)")

data = pd.DataFrame({
    "pixel": patch_seq,
})

# sns.set(rc={"figure.figsize": (6, 3)})
# fig, ax = plt.subplots(figsize=(6, 4))
g = sns.displot(data=data, x='pixel')
g.fig.set_size_inches(6,3)
plt.xlabel("Grayscale value")
plt.savefig("grascale_value.png", bbox_inches='tight',)
# hist, idx = histogram(patch, nbins=256)
# print(hist, idx)
```
