```python
%matplotlib inline
import matplotlib.pylab as plt
plt.style.use('ggplot')
import seaborn as sns
import pandas as pd
import numpy as np


# sns.set(rc={"figure.figsize":(5.4, 3)}) 
fig = plt.figure(figsize=(8, 3))
data = np.load("../cache/optimal_brct_im.npy", allow_pickle=True) # from search_optimal_ceph.py
data = data.tolist()
# print(data)
df = pd.DataFrame.from_dict(data)
df['mre_gap'] = df['mre_lm'] - df["mre_lm2"]
print(df)
xlabel = "cj_ct_floor"
ylabel = "cj_br_floor"
fig = sns.scatterplot(x=xlabel, y=ylabel, data=df)
sns.regplot(x=xlabel, y=ylabel, data=df)


sns.scatterplot(x=xlabel, y=ylabel, data=df[7:8], color="red")
sns.scatterplot(x=xlabel, y=ylabel, data=df[17:18], color="black")
print(df[0:1])

# scatter_fig = fig.get_figure()
# scatter_fig.savefig("./tmp/_search_optimal_ceph.png")
```

```python
xlabel = "entr_mean"
# ylabel = "mi_mean"
ylabel = "cj_br_ceil"
fig = plt.figure(figsize=(4*1.8, 4))
plt.xlim([4.0,5.1])
fig = sns.scatterplot(x=xlabel, y=ylabel, data=df)
sns.regplot(x=xlabel, y=ylabel, data=df)
fig.set_xlabel("Image Information Entropy (IIE)")
fig.set_ylabel("Augmentation Intensity")
plt.savefig("entr-aug.pdf")
```
