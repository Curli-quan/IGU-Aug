```python
%matplotlib inline
import matplotlib.pylab as plt
plt.style.use('ggplot')
import seaborn as sns
import pandas as pd
import palettable

#python palettable库

# plt.figure(figsize=(4, 4)) 
fig, ax = plt.subplots(1,2, figsize=(9,4))
sns.palplot(sns.color_palette(palette=palettable.scientific.sequential.Nuuk_7.mpl_colors))
sns.palplot(sns.cubehelix_palette(8, start=0.5, rot=-0.5))

# for i in range(12):
plt.sca(ax[0])
for i in range(12):
    data = pd.read_csv(f"../mi_vs_aug_lmid_{i}.csv")
    sns.lineplot(data=data, x='bright', y='mi', ax=ax[0], color=sns.cubehelix_palette(16, start=0.5, rot=-0.5)[2])
# sns.lineplot(data=data, x='bright', y='mi', ax=ax[1], label="Cephalometric", color=sns.cubehelix_palette(16, start=0.5, rot=-0.5)[2])

for i in range(10):
    data = pd.read_csv(f"../mi_vs_aug_lmid_{i}_hand.csv")
    sns.lineplot(data=data, x='bright', y='mi', ax=ax[0], color=sns.cubehelix_palette(16, start=0.5, rot=-0.5)[8])
# sns.lineplot(data=data, x='bright', y='mi', label="HandXray", ax=ax[1], color=sns.cubehelix_palette(16, start=0.5, rot=-0.5)[7])

plt.xlabel("Brightness")
plt.ylabel("Mutual Information (MI)")
plt.tight_layout()

plt.sca(ax[1])
for i in range(16):
    data = pd.read_csv(f"../mi_vs_aug2_lmid_{i}.csv")
    sns.lineplot(data=data, x='contrast', y='mi', ax=ax[1], color=sns.cubehelix_palette(16, start=0.5, rot=-0.5)[2])
sns.lineplot(data=data, x='contrast', y='mi', ax=ax[1], label="Cephalometric", color=sns.cubehelix_palette(16, start=0.5, rot=-0.5)[2])

for i in range(10):
    data = pd.read_csv(f"../mi_vs_aug2_lmid_{i}_hand.csv")
    sns.lineplot(data=data, x='contrast', y='mi', ax=ax[1], color=sns.cubehelix_palette(16, start=0.5, rot=-0.5)[8])
sns.lineplot(data=data, x='contrast', y='mi', label="HandXray", ax=ax[1], color=sns.cubehelix_palette(16, start=0.5, rot=-0.5)[8])
plt.xlabel("Contrast")
plt.ylabel("")

plt.savefig("mi_aug2.pdf")
```
