```python
%matplotlib inline
# RegPlot
import matplotlib.pylab as plt
plt.style.use('ggplot')
import seaborn as sns
import pandas as pd
import numpy as np
from einops import rearrange

y = np.array([23.35, 23.28, 23.76, 23.87, 23.43, 23.4, 23.58])
y_list = np.array([y + np.random.random(7) - 0.5 for _ in range(10)])
yy = rearrange(y_list, "h w -> (h w)")
x = np.array([1,2,3,4,5,6,7]) - 0.5
x_list = np.array([x + np.random.random(7) * 0.1 - 0.05  for _ in range(10)])
xx = rearrange(x_list, "h w -> (h w)")

sns.regplot(x=xx, y=yy, order=4)
```
