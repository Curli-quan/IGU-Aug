```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
plt.style.use('ggplot')

pd_data = pd.read_csv("/home1/quanquan/code/landmark/code/tproj//loss_iie_hand.csv")
colors = ["#002c53", "#ffa510", "#0c84c6", "#ffbd66", "#f74d4d", "#2455a4", "#41b7ac"]
ax = sns.regplot(data=pd_data, x="entr", y="loss", marker='*', order=3, scatter_kws={'s': 40, 'color': colors[0]},line_kws={'linestyle':'--', 'color': colors[2]}) # 'color':'#016392',
plt.ylim(23.3,25.65)
plt.xlim(1.5,6)
y = pd_data['loss'].mean()
print(y.mean())
# scatters = pd.DataFrame({
#     "x": [2.42, 4.42],
#     "y": [23.656483, 23.656483],
# })
plt.hlines([y], 1.5,6, linestyles='dashed', colors=colors[6], label='Avg. loss')
# sns.scatterplot(data=scatters, x="x", y="y", color='green')
# plt.plot([2.42, 4.42],[23.656483, 23.656483],)
plt.plot([3.4133, 5.72],[y, y],'o', color=colors[1])

# plt.vlines([2.42], 23, 24, linestyles='dashed', colors='blue')
# plt.vlines([4.42], 23, 24, linestyles='dashed', colors='green')
# data2
plt.text(x=1.66, y=24.7, s=r"Avg. ",  fontdict={"size":16, "color":colors[6]})
plt.text(x=1.66, y=24.48, s=r"loss",  fontdict={"size":16, "color":colors[6]})
plt.xlabel("Image Information Entropy (IIE)")
plt.ylabel("InfoNCE Loss")
plt.tight_layout()
plt.savefig("./loss_iie2.pdf")
```
