```python
%matplotlib inline
import matplotlib.pylab as plt
import pandas as pd
import seaborn as sns
plt.style.use('ggplot')

df = pd.DataFrame({
    "perc": ["0 %","5 %","10 %","25 %","0 %","5 %","10 %","25 %",],
    "mre": [2.84, 2.91 ,2.95, 3.01, 2.35,2.77,2.93,2.98],
    "comparison": ["baseline", "baseline", "baseline", "baseline", "ours", "ours","ours","ours", ],
})
plt.figure(figsize=(6,4))
sns.barplot(df, x="perc", y="mre", hue="comparison")
plt.ylim(2,3.1)
plt.ylabel("MRE")
plt.xlabel("Noisy images")
plt.tight_layout()
plt.savefig("noise_plot.pdf")
```

```python
%matplotlib inline
import matplotlib.pylab as plt
import pandas as pd
import seaborn as sns
plt.style.use('ggplot')

sns.set_style("whitegrid",{"font.sans-serif":['SimHei']})
df = pd.DataFrame({
    "perc": ["0 %","5 %","10 %","25 %","0 %","5 %","10 %","25 %",],
    "mre": [2.84, 2.91 ,2.95, 3.01, 2.35,2.77,2.93,2.98],
    "比较": ["基线方法", "基线方法", "基线方法", "基线方法", "本文方法", "本文方法","本文方法","本文方法", ],
})
plt.figure(figsize=(6,4))
sns.barplot(df, x="perc", y="mre", hue="比较")
plt.ylim(2,3.1)
plt.ylabel("MRE")
plt.xlabel("噪音数据的占比")
plt.tight_layout()
plt.savefig("noise_plot_cn.pdf")
```
