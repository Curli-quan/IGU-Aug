```python
%matplotlib inline
import matplotlib.pylab as plt
import seaborn as sns
import pandas as pd
plt.style.use('ggplot')
import numpy as np

# The data copied from search_aug.ipynb
mi_mean = [0.58855864,0.9689842,0.88423108,0.41595766,0.93467499,0.87520715
            ,0.75355031,0.69097756,0.70706193,0.9617875,0.95781572,0.96094506
            ,0.98639255,0.89073955,0.96499311,0.71407015,0.93329412,0.91891464
            ,0.7271314]
m = np.array(mi_mean).mean()
print(m)

fig = plt.figure(dpi=100, figsize=(5,3))
data =pd.read_csv("../mre_aug2.csv")
data["I_data"] = data['alpha'] * m
print(data)
sns.lineplot(data=data, x="I_data", y="mre", color='#375C9E')
plt.xlabel(r"$\mu(I_{aug})$")
plt.ylabel("MRE")
plt.savefig("mre_aug_alpha.pdf")
plt.vlines([m+0.01], 2.3,2.45, colors="#006814", linestyles='dashed', label='Ours')
plt.vlines([m], 2.3,2.45, colors="#9D0202", linestyles='dashed', label=r'$\mu(I_{data})$')
plt.vlines([m*1.7], 2.3,2.45, colors="grey", linestyles='dashed', label='Baseline')

plt.text(m*1.7-0.12, 2.46, s="Baseline")
plt.text(m+0.03, 2.43, s="Ours", color="#006814")
plt.text(m-0.175, 2.43, s='$\mu(I_{data})$', color="#9D0202")

# plt.legend()
# plt.tight_layout()
plt.savefig("mre_mu_aug.pdf",bbox_inches='tight',)
```
